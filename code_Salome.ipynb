```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Charger le fichier CSV
donnees_csv = pd.read_csv("student-por.csv")

# Créer une matrice à partir des données CSV
matrice = np.array(donnees_csv.values)

n, p = matrice.shape

reason = matrice[:,10] #quali
guardian = matrice[:,11] #quali
traveltime = matrice[:,12] #quali
studytime = matrice[:,13] #qualii
failures = matrice[:,14] #quanti
schoolsup = matrice[:,15] #quali
famsup = matrice[:,16] #quali
paid = matrice[:,17] #quali
activities = matrice[:,18] #quali
nursery = matrice[:,19] #quali

plt.figure(figsize=(20,20))

#reason
modalite_reason, eff_reason = np.unique(reason, return_counts = True)
plt.subplot(4,4,1)
plt.pie(eff_reason, labels=modalite_reason, autopct = '%1.1f%%');
plt.title('Reason')


#guardian
modalite_guardian, eff_guardian = np.unique(guardian, return_counts = True)
plt.subplot(4,4,2)
plt.pie(eff_guardian, labels=modalite_guardian, autopct = '%1.1f%%');
plt.title('Guardian')


#traveltime
modalite_Ttrajet, eff_Ttrajet = np.unique(traveltime, return_counts =True)
plt.subplot(4,4,3)
plt.pie(eff_Ttrajet, labels = modalite_Ttrajet, autopct = '%1.1f%%');
plt.title('Traveltime')


#studytime
modalite_Ttravail, eff_Ttravail = np.unique(studytime, return_counts =True)
plt.subplot(4,4,4)
plt.pie(eff_Ttravail, labels = modalite_Ttravail, autopct = '%1.1f%%');
plt.title('Studytime')


#failures
modalite_failures, eff_failures = np.unique(failures, return_counts =True)
plt.subplot(4,4,5)
plt.pie(eff_failures, labels = modalite_failures, autopct = '%1.1f%%');
plt.title('Failures')


#schoolsup
modalite_schoolsup, eff_schoolsup = np.unique(schoolsup, return_counts =True)
plt.subplot(4,4,6)
plt.pie(eff_schoolsup, labels = modalite_schoolsup, autopct = '%1.1f%%');
plt.title('SchoolSup')


#famsup
modalite_famsup, eff_famsup = np.unique(famsup, return_counts =True)
plt.subplot(4,4,7)
plt.pie(eff_famsup, labels = modalite_famsup, autopct = '%1.1f%%');
plt.title('FamSup')


#paid
modalite_paid, eff_paid = np.unique(paid, return_counts =True)
plt.subplot(4,4,8)
plt.pie(eff_paid, labels = modalite_paid, autopct = '%1.1f%%');
plt.title('Paid')


#activities
modalite_activities, eff_activities = np.unique(activities, return_counts =True)
plt.subplot(4,4,9)
plt.pie(eff_activities, labels = modalite_activities, autopct = '%1.1f%%');
plt.title('Activities')


#nursery
modalite_nursery, eff_nursery = np.unique(nursery, return_counts =True)
plt.subplot(4,4,10)
plt.pie(eff_nursery, labels = modalite_nursery, autopct = '%1.1f%%');
plt.title('Nursery')
```
